## ทฤษฎีความพึงพอใจของลูกค้า (CSAT)

ความพึงพอใจของลูกค้า (CSAT) เป็นตัวบ่งชี้ประสิทธิภาพหลักที่ใช้กันทั่วไปซึ่งติดตามความพึงพอใจของลูกค้าที่มีต่อผลิตภัณฑ์และ / หรือบริการขององค์กร เป็นสถานะทางจิตวิทยาที่วัดจากความคาดหวังของลูกค้า เมื่อทราบว่าความคาดหวังเหล่านั้นคืออะไรคุณสามารถเพิ่มความภักดีของลูกค้าให้กับแบรนด์ของคุณได้อย่างมาก



## ความพึงพอใจคืออะไร?

ความพึงพอใจเป็นสถานะทางจิตวิทยาโดยรวมที่สะท้อนถึงการประเมินความสัมพันธ์ระหว่างลูกค้าหรือผู้บริโภคกับ บริษัท สิ่งแวดล้อมผลิตภัณฑ์หรือบริการ ความพึงพอใจเกี่ยวข้องกับองค์ประกอบทางจิตวิทยาหนึ่งในสามองค์ประกอบดังต่อไปนี้: ความรู้ความเข้าใจ (การคิด / การประเมิน) อารมณ์ (อารมณ์ / ความรู้สึก) และพฤติกรรม


## ความคาดหวังและความพึงพอใจของลูกค้า

ความคาดหวังคือความเชื่อ (ความน่าจะเป็นหรือความน่าจะเป็น) ที่ผลิตภัณฑ์และ / หรือบริการ (ที่มีคุณลักษณะคุณลักษณะหรือคุณลักษณะบางอย่าง) จะสร้างผลลัพธ์ที่แน่นอน (ประโยชน์หรือคุณค่า) จากระดับประสิทธิภาพที่คาดการณ์ไว้ ความคาดหวังมักเกี่ยวข้องกับความพึงพอใจและสามารถวัดได้ดังนี้

1. ความสำคัญ: คุณค่าของผลิตภัณฑ์ / บริการตอบสนองความคาดหวัง
2. ความคาดหวังโดยรวมที่มีผลต่อความพึงพอใจ: ชอบ / ไม่ชอบสินค้า / บริการ
3. การปฏิบัติตามความคาดหวัง: ระดับประสิทธิภาพที่คาดหวังกับความคาดหวังที่ต้องการ นี่คือ“ การบรรลุเป้าหมายที่คาดการณ์ได้” และเป็นดัชนีเฉพาะของผู้ตอบแบบสอบถามในระดับประสิทธิภาพที่จำเป็นต่อการตอบสนอง
4. มูลค่าที่คาดหวังจากการใช้งาน: ความพึงพอใจมักถูกกำหนดโดยความถี่ของการใช้งาน หากไม่ใช้ผลิตภัณฑ์ / บริการบ่อยเท่าที่ควรผลลัพธ์อาจไม่เป็นที่พอใจ ตัวอย่างเช่นรถจักรยานยนต์ที่อยู่ในโรงรถการสมัครสมาชิกศูนย์ออกกำลังกายในท้องถิ่นหรือการใช้บัตรผ่านฤดูกาลที่สกีรีสอร์ทจะทำให้เกิดความไม่พอใจต่อการตัดสินใจซื้อมากกว่าการใช้ผลิตภัณฑ์ / บริการจริง

## การวัดความคาดหวัง

ในการสร้างแบบสำรวจความพึงพอใจของลูกค้าจะเป็นประโยชน์ในการพิจารณาเหตุผลที่ความคาดหวังก่อนซื้อหรือความพึงพอใจหลังการซื้ออาจไม่สมบูรณ์หรือวัดได้

1. ความคาดหวังอาจไม่สะท้อนถึงคุณลักษณะของบริการที่ไม่คาดคิด
2. ความคาดหวังอาจจะค่อนข้างคลุมเครือสร้างละติจูดกว้างของการยอมรับในประสิทธิภาพและความพึงพอใจที่คาดหวัง
3. การประเมินความคาดหวังและประสิทธิภาพของผลิตภัณฑ์อาจมีความรู้สึกและไม่รับรู้เช่นเดียวกับในด้านรสนิยมสไตล์หรือภาพ
4. การใช้ผลิตภัณฑ์อาจดึงดูดความสนใจเพียงเล็กน้อยจนไม่ก่อให้เกิดผลกระทบอย่างมีสติหรือการรับรู้ (การประเมิน) และผลลัพธ์ในการวัดความพึงพอใจหรือความไม่พอใจ
5. อาจมีประโยชน์ที่ไม่คาดคิดหรือผลที่ตามมาจากการซื้อหรือการใช้ผลิตภัณฑ์ (เช่นการใช้งานหรือคุณสมบัติที่ไม่คาดว่าจะมีการซื้อ)
6. ความคาดหวังดั้งเดิมอาจสูงหรือต่ำเกินจริง
7. ผู้ซื้อผลิตภัณฑ์ ผู้มีอิทธิพล และผู้ใช้อาจเป็นบุคคลที่แตกต่างกันซึ่งแต่ละคนมีความคาดหวังที่แตกต่างกัน


## วิธีการวัดความพึงพอใจของลูกค้า: 4 ตัวชี้วัดหลัก

มีกลยุทธ์มากมาย แต่การมองเห็นพื้นฐานของวิธีการวัดความพึงพอใจของลูกค้าที่ไม่ดีอาจเป็นอันตรายต่อธุรกิจ ต่อไปนี้คือการวัดความพึงพอใจของลูกค้า 4 ประการที่มีความสำคัญต่อความสำเร็จของธุรกิจของคุณ

1. Overall Satisfaction Measure (Attitudinal)

ความพึงพอใจโดยรวมสะท้อนความเห็นโดยรวมของประสบการณ์ความพึงพอใจของผู้บริโภคต่อผลิตภัณฑ์ที่เขาใช้ และตัวทำนายที่ดีที่สุดเพียงอย่างเดียวเกี่ยวกับความพึงพอใจของลูกค้าคือประสบการณ์ของลูกค้าที่ส่งผลให้เกิดการอ้างถึงคุณภาพ

การรับรู้คุณภาพมักถูกวัดในหนึ่งในสามบริบท:

    1) คุณภาพโดยรวม
    2) การรับรู้ความน่าเชื่อถือ
    3) ครอบคลุมความต้องการของลูกค้าเป็นจริง

2. Loyalty Measurement (Affective, Behavioral)

ความภักดีของลูกค้าสะท้อนถึงโอกาสในการซื้อคืนผลิตภัณฑ์หรือบริการ ความพึงพอใจของลูกค้าเป็นตัวทำนายหลักของการซื้อคืน แต่ได้รับอิทธิพลอย่างมากจากการประเมินประสิทธิภาพที่ชัดเจนของประสิทธิภาพของผลิตภัณฑ์คุณภาพและคุณค่า

ความภักดีมักวัดจากการรวมกันของมาตรการรวมถึงความพึงพอใจโดยรวมโอกาสในการซื้อคืนและโอกาสในการแนะนำแบรนด์ให้เพื่อน

โดยทั่วไปความภักดีอาจเป็นคะแนนรวมสำหรับคำถามสามข้อต่อไปนี้:

    1) โดยรวมแล้วคุณพอใจกับ [brand] แค่ไหน
    2) คุณมีแนวโน้มที่จะเลือก / กลับมาซื้อ [brand] ต่อไปอย่างไร
    3) มีโอกาสมากที่คุณจะแนะนำ [แบรนด์] ให้กับเพื่อนหรือสมาชิกในครอบครัวหรือไม่
    
3. A Series of Attribute Satisfaction Measurements (Affective and Cognitive)

ผลกระทบที่แสดงออกมา (liking / disliking) เป็นตัววัดที่ดีที่สุดในบริบทของคุณสมบัติหรือประโยชน์ของผลิตภัณฑ์ ความพึงพอใจของลูกค้าได้รับอิทธิพลจากการรับรู้คุณภาพของผลิตภัณฑ์และคุณลักษณะของบริการและมีการกลั่นกรองโดยความคาดหวังของผลิตภัณฑ์หรือบริการ ผู้วิจัยจะต้องกำหนดและพัฒนามาตรการสำหรับแต่ละคุณลักษณะที่มีความสำคัญต่อความพึงพอใจของลูกค้า

ทัศนคติของผู้บริโภคที่มีต่อผลิตภัณฑ์ที่พัฒนาขึ้นอันเป็นผลมาจากข้อมูลผลิตภัณฑ์หรือประสบการณ์ใด ๆ กับผลิตภัณฑ์ไม่ว่าจะรับรู้หรือจริง

การวัดทัศนคติต่อผลิตภัณฑ์หรือบริการที่ผู้บริโภคไม่เคยใช้ ไม่มีความหมายในการวัดความพึงพอใจเมื่อไม่มีการใช้ผลิตภัณฑ์หรือบริการ

ความรู้ความเข้าใจหมายถึงการตัดสิน: ผลิตภัณฑ์มีประโยชน์ (หรือไม่เป็นประโยชน์); เหมาะสมกับสถานการณ์ (หรือไม่พอดี); เกินความต้องการของปัญหา / สถานการณ์ (หรือไม่เกิน); หรือเป็นส่วนสำคัญของประสบการณ์ผลิตภัณฑ์ (หรือไม่สำคัญ)

การตัดสินมักจะเฉพาะสิ่งที่ตั้งใจและโอกาสในการใช้ผลิตภัณฑ์ที่ซื้อมา

ผลกระทบและความพึงพอใจ แสดงถึงผลกระทบทางอารมณ์ที่เกิดจากการใช้ผลิตภัณฑ์และคุณภาพของสินค้า

4. Intentions to Repurchase Measurements (Behavioral Measures)

อนาคตหรือพฤติกรรมผู้บริโภค สะท้อนถึงประสบการณ์ที่ผ่านมาของผู้บริโภค

ความพึงพอใจสามารถมีอิทธิพลต่อการกระทำหลังการซื้อ เช่นการสื่อสารกับผู้อื่นผ่านทางคำพูดจากปากและเครือข่ายสังคม

การดำเนินการหลังประสบการณ์เพิ่มเติม อาจสะท้อนถึงระดับการมีส่วนร่วมของผลิตภัณฑ์ที่สูงขึ้น ซึ่งส่งผลให้มีการค้นหาผลิตภัณฑ์หรือข้อมูลเพิ่มขึ้น ลดการทดลองใช้ผลิตภัณฑ์ทางเลือกลดลง

# 1. Data set
The dataset from customer satisfaction survey here : https://www.kaggle.com/najibmh/us-airline-passenger-satisfaction-survey

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('Satisfaction.csv')
df.head()
```

```python
df.columns.values
```

# 2. Use mean 

### 2.1 Find satisfaction on each attribute

```python
# Average score of each customer
df_score = pd.DataFrame(df[['id', 'satisfaction_v2', 'Gender', 'Customer Type', 'Age',
       'Type of Travel', 'Class', 'Flight Distance', 'Departure Delay in Minutes',
       'Arrival Delay in Minutes','Seat comfort',
       'Departure/Arrival time convenient', 'Food and drink',
       'Gate location', 'Inflight wifi service', 'Inflight entertainment',
       'Online support', 'Ease of Online booking', 'On-board service',
       'Leg room service', 'Baggage handling', 'Checkin service',
       'Cleanliness', 'Online boarding']])

df_score['Average Satisfaction'] = df_score.iloc[:, 10:].mean(axis = 1) 
df_score
```

### 2.2 Satisfaction Score

```python
print ('Summary Satisfaction')
satisfaction_table = pd.DataFrame(df_score.iloc[:, 10:].mean(axis = 0))
satisfaction_table.columns = ['Score']
satisfaction_table['SD'] = df_score.iloc[:, 10:].std(axis = 0)
satisfaction_table
```

### 2.3 Visualization

```python
# Create dataset
height = satisfaction_table['Score']
bars = satisfaction_table.index.values
y_pos = np.arange(len(bars))

# Create bars
plt.bar(y_pos, height)

# Create names on the x-axis
plt.xticks(y_pos, bars, size = 4, rotation = 45)

plt.show()
```
